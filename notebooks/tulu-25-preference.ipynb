```python
from datasets import load_dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# Load a specific split (e.g., alpaca_farm gpt4 and human)
dataset_gpt4 = load_dataset("allenai/tulu-2.5-preference-data", 
                            split="alpaca_farm_gpt4_pref")

df_gpt4 = dataset_gpt4.to_pandas()

dataset_human = load_dataset("allenai/tulu-2.5-preference-data", 
                             split="alpaca_farm_human_pref")

df_human = dataset_human.to_pandas()
```

### Get some basic statistics about data and responses

```python
# Basic statistics
print(f"GPT-4 split: {len(dataset_gpt4)} examples")
print(f"Human split: {len(dataset_human)} examples")
print(f"Total examples: {len(df_gpt4) + len(df_human)}")

# # Check structure
# print(f"GPT4: {df_gpt4.head()}\n{df_gpt4.columns}")
# print(f"Human: {df_human.head()}\n{df_human.columns}")

# Response length distributions

def extract_length(array):
    try:
        return len(array[-1]['content'])
    except:
        return 0

# First gpt4 chosen and rejected
df_gpt4['chosen_length'] = df_gpt4['chosen'].apply(
    extract_length
)   
df_gpt4['rejected_length'] = df_gpt4['rejected'].apply(
    extract_length
)
print("GPT-4 Response Lengths:")
print(df_gpt4[['chosen_length', 'rejected_length']].describe())

# Then human chosen and rejected
df_human['chosen_length'] = df_human['chosen'].apply(
    extract_length
)
df_human['rejected_length'] = df_human['rejected'].apply(
    extract_length
)  
print("Human Response Lengths:")
print(df_human[['chosen_length', 'rejected_length']].describe())

```
